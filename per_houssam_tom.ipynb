```python
def dual(f,fq):
    g = copy(f)
    ligneg = len(g.rows())
    colg = len(g.columns())
    carreg = matrix(fq,[[g[i,j] for j in range(ligneg)] for i in range(ligneg)])
    invcar = carreg**-1
    gid = invcar*g
    d = matrix(fq,[[0 for a in range(colg)] for b in range(colg-ligneg)])
    ide = identity_matrix(fq,colg-ligneg)
    for i in range(colg-ligneg):
        for j in range(ligneg):
            d[i,j] = -gid[j][i+ligneg]
        for k in range(ligneg,colg):
            d[i,k] = ide[i,k-ligneg]
    return d
```

```python
def Decalage(l):
    return [l[len(l)-1]] + l[:len(l)-1]
```

```python
def MatGen(la,lb):
    taille = len(la)
    a = matrix(GF(2),taille,taille)
    b = matrix(GF(2),taille,taille)
    for i in range(taille):
        for j in range(taille):
            a[i,j] = la[j]
            b[i,j] = lb[j]
        la = Decalage(la)
        lb = Decalage(lb)
    Sk = matrix(GF(2),taille,taille*2)
    for i in range(taille):
        for j in range(taille):
            Sk[i,j] = a[i,j]
        for k in range(taille,taille*2):
            Sk[i,k] = b[i,k-taille]
    return a,Sk
```

```python
def qualite_test(n,seuil,tour,erreur,Pk):
    r = 0
    d = 0
    #mot = [0 for b in range (1024)]
    for i in range (n):
        print(r,d)
        mot = Pk[randint(0,511)]
        for k in range(10):
            mot += Pk[randint(0,511)]
        mot2 = copy(mot)
        for j in range(erreur):
            mot[randint(0,1023)] += 1
        chiffre = matrix(GF(2),mot)
        message = bitflipv3(H,chiffre,seuil,tour)
        if message == "erreur" :
            d = d+1
        if message == mot2 :
            r = r+1
    return ("pour un seuil de ",seuil, "on a tant de réussite",r,"et d'échec",d)
```
