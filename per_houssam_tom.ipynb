```python
def dual(f,fq):
    g = copy(f)
    ligneg = len(g.rows())
    colg = len(g.columns())
    carreg = matrix(fq,[[g[i,j] for j in range(ligneg)] for i in range(ligneg)])
    invcar = carreg**-1
    gid = invcar*g
    d = matrix(fq,[[0 for a in range(colg)] for b in range(colg-ligneg)])
    ide = identity_matrix(fq,colg-ligneg)
    for i in range(colg-ligneg):
        for j in range(ligneg):
            d[i,j] = -gid[j][i+ligneg]
        for k in range(ligneg,colg):
            d[i,k] = ide[i,k-ligneg]
    return d
```

```python
def Decalage(l):
    return [l[len(l)-1]] + l[:len(l)-1]
```

```python
def MatGen(la,lb):
    taille = len(la)
    a = matrix(GF(2),taille,taille)
    b = matrix(GF(2),taille,taille)
    for i in range(taille):
        for j in range(taille):
            a[i,j] = la[j]
            b[i,j] = lb[j]
        la = Decalage(la)
        lb = Decalage(lb)
    Sk = matrix(GF(2),taille,taille*2)
    for i in range(taille):
        for j in range(taille):
            Sk[i,j] = a[i,j]
        for k in range(taille,taille*2):
            Sk[i,k] = b[i,k-taille]
    return a,Sk
```
